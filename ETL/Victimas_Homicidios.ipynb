***
### Crear columnas de fecha desglosada y edad cambiar el tipo Variable

```python
# Convertir columna en Variable tiempo paa tratamiento
victimas['FECHA_FALLECIMIENTO'] = pd.to_datetime(victimas['FECHA_FALLECIMIENTO'], errors='coerce', format='%Y-%m-%d', utc=False)

# Crear nuevas columnas para año, mes 
victimas['Año_Fallecimiento'] = victimas['FECHA_FALLECIMIENTO'].dt.year
victimas['Mes_Fallecimiento'] = victimas['FECHA_FALLECIMIENTO'].dt.month
victimas['Dia_Fallecimiento'] = victimas['FECHA_FALLECIMIENTO'].dt.day

# Remplaza NaT por "SD"
victimas['FECHA_FALLECIMIENTO'] = victimas['FECHA_FALLECIMIENTO'].fillna("SD")

# Muestra el DataFrame resultante
print(victimas[['FECHA_FALLECIMIENTO', 'Año_Fallecimiento', 'Mes_Fallecimiento', 'Dia_Fallecimiento']])
```

```python
# Cambia el tipo de Varible de las columnas
victimas['Año_Fallecimiento'] = pd.to_numeric(victimas['Año_Fallecimiento'], errors='coerce')
victimas['Mes_Fallecimiento'] = pd.to_numeric(victimas['Mes_Fallecimiento'], errors='coerce')
victimas['Dia_Fallecimiento'] = pd.to_numeric(victimas['Dia_Fallecimiento'], errors='coerce')
victimas['FECHA_FALLECIMIENTO'] = pd.to_datetime(victimas['FECHA_FALLECIMIENTO'], format='%Y-%m-%d %H:%M:%S', errors='coerce')
```
