### Generar una funcion para dividir el año y crear una columna 'Semestre'

```python
# Definir una función para asignar el número del semestre
def asignar_semestre(mes):
    if 1 <= mes <= 6:
        return 1
    elif 7 <= mes <= 12:
        return 2
    else:
        return None

# Aplicar la función a la columna 'MM' para crear la columna 'semestre'
kpi['Semestre'] = kpi['Mes_Hechos'].apply(asignar_semestre)

# Mostrar el DataFrame resultante
kpi    
```
